```python
import pandas as pd
import holidays

# Load your data into a DataFrame
# data_combined = pd.read_csv('path_to_your_combined_data.csv')  # Adjust path as needed

# Define function to get holiday name based on location
def get_holiday_name(date, location):
    year = date.year
    if location == 'Bucharest':
        ro_holidays = holidays.RO(years=year)
        return ro_holidays.get(date, None)
    elif location == 'Los Angeles':
        us_holidays = holidays.US(years=year)
        return us_holidays.get(date, None)
    elif location == 'Madrid':
        es_holidays = holidays.ES(years=year)
        return es_holidays.get(date, None)
    elif location == 'Nairobi':
        ke_holidays = holidays.KE(years=year)
        return ke_holidays.get(date, None)
    elif location == 'Rome':
        it_holidays = holidays.IT(years=year)
        return it_holidays.get(date, None)
    else:
        return None

# Define function to get the day type description
def get_day_description(date, holiday_name):
    day_of_week = date.day_name()
    if holiday_name:
        return holiday_name
    elif day_of_week in ['Saturday', 'Sunday']:
        return 'Weekend Day'
    else:
        return 'Working Day'

# Apply the function to add the holiday name column
def add_holiday_column(df):
    df['HolidayName'] = df.apply(lambda row: get_holiday_name(row['DateLocal'], row['location_name']), axis=1)
    return df

# Apply the function to add the day description column
def add_day_description_column(df):
    df['DayDescription'] = df.apply(lambda row: get_day_description(row['DateLocal'], row['HolidayName']), axis=1)
    return df

# Ensure DateLocal is in datetime format
data_combined_sorted['DateLocal'] = pd.to_datetime(data_combined_sorted['DateLocal'])

# Add the holiday name column and the day description column
data_combined_sorted = add_holiday_column(data_combined_sorted)
data_combined_sorted = add_day_description_column(data_combined_sorted)

# Check the updated DataFrame
print(data_combined_sorted.head())
```

```python
import pandas as pd

# Assuming data_combined_sorted is your DataFrame
# List of pollutants
pollutants = ['O3', 'PM10', 'PM2.5', 'NO2', 'CO', 'NO', 'NOx', 'SO2']

# Define city-specific pollutant availability
city_pollutants = {
    'Nairobi': ['PM2.5'],
    'Rome': ['NO', 'NOx', 'O3', 'PM10', 'PM2.5', 'NO2'],
    'Madrid': ['O3', 'PM10', 'PM2.5', 'NO2'],
    'Los Angeles': ['NOx', 'O3', 'PM2.5', 'NO2', 'NO'],
    'Bucharest': ['PM10', 'NO2', 'SO2', 'CO']
}

# Function to clean pollutants based on city-specific availability
def clean_pollutants(row, city_pollutants):
    city = row['location_name']
    available_pollutants = city_pollutants[city]
    for pollutant in pollutants:
        if pollutant not in available_pollutants:
            row[pollutant] = 0
    return row

# Apply the cleaning function
data_combined_sorted = data_combined_sorted.apply(clean_pollutants, city_pollutants=city_pollutants, axis=1)

# Print the cleaned DataFrame to verify
print("Cleaned DataFrame:")
print(data_combined_sorted.head())
```

```python
import pandas as pd
import numpy as np

# Define the linear scaling function
def linear_scale(value, old_min, old_max, new_min, new_max):
    return ((value - old_min) * (new_max - new_min) / (old_max - old_min)) + new_min

# Define AQI calculation functions for different pollutants
def calculate_aqi_pm25(value):
    if value <= 12.0:
        return linear_scale(value, 0, 12, 0, 50)
    elif value <= 35.4:
        return linear_scale(value, 12.1, 35.4, 51, 100)
    elif value <= 55.4:
        return linear_scale(value, 35.5, 55.4, 101, 150)
    elif value <= 150.4:
        return linear_scale(value, 55.5, 150.4, 151, 200)
    elif value <= 250.4:
        return linear_scale(value, 150.5, 250.4, 201, 300)
    elif value <= 350.4:
        return linear_scale(value, 250.5, 350.4, 301, 400)
    elif value <= 500.4:
        return linear_scale(value, 350.5, 500.4, 401, 500)
    else:
        return 500  # AQI value capped at 500 for PM2.5

def calculate_aqi_o3(value):
    if value <= 54:
        return linear_scale(value, 0, 54, 0, 50)
    elif value <= 70:
        return linear_scale(value, 55, 70, 51, 100)
    elif value <= 85:
        return linear_scale(value, 71, 85, 101, 150)
    elif value <= 105:
        return linear_scale(value, 86, 105, 151, 200)
    elif value <= 200:
        return linear_scale(value, 106, 200, 201, 300)
    elif value <= 504:
        return linear_scale(value, 201, 504, 301, 400)
    else:
        return 500  # AQI value capped at 500 for O3

def calculate_aqi_no2(value):
    if value <= 53:
        return linear_scale(value, 0, 53, 0, 50)
    elif value <= 100:
        return linear_scale(value, 54, 100, 51, 100)
    elif value <= 360:
        return linear_scale(value, 101, 360, 101, 150)
    elif value <= 649:
        return linear_scale(value, 361, 649, 151, 200)
    elif value <= 1249:
        return linear_scale(value, 650, 1249, 201, 300)
    elif value <= 2049:
        return linear_scale(value, 1250, 2049, 301, 400)
    else:
        return 500  # AQI value capped at 500 for NO2

def calculate_aqi_pm10(value):
    if value <= 54:
        return linear_scale(value, 0, 54, 0, 50)
    elif value <= 154:
        return linear_scale(value, 55, 154, 51, 100)
    elif value <= 254:
        return linear_scale(value, 155, 254, 101, 150)
    elif value <= 354:
        return linear_scale(value, 255, 354, 151, 200)
    elif value <= 424:
        return linear_scale(value, 355, 424, 201, 300)
    elif value <= 504:
        return linear_scale(value, 425, 504, 301, 400)
    else:
        return 500  # AQI value capped at 500 for PM10

def calculate_aqi_co(value):
    if value <= 4.4:
        return linear_scale(value, 0, 4.4, 0, 50)
    elif value <= 9.4:
        return linear_scale(value, 4.5, 9.4, 51, 100)
    elif value <= 12.4:
        return linear_scale(value, 9.5, 12.4, 101, 150)
    elif value <= 15.4:
        return linear_scale(value, 12.5, 15.4, 151, 200)
    elif value <= 30.4:
        return linear_scale(value, 15.5, 30.4, 201, 300)
    elif value <= 40.4:
        return linear_scale(value, 30.5, 40.4, 301, 400)
    elif value <= 50.4:
        return linear_scale(value, 40.5, 50.4, 401, 500)
    else:
        return 500  # AQI value capped at 500 for CO

def calculate_aqi_no(value):
    return 0  # Placeholder, as NO typically isn't directly converted to AQI

def calculate_aqi_nox(value):
    return 0  # Placeholder, as NOx typically isn't directly converted to AQI

def calculate_aqi_so2(value):
    if value <= 35:
        return linear_scale(value, 0, 35, 0, 50)
    elif value <= 75:
        return linear_scale(value, 36, 75, 51, 100)
    elif value <= 185:
        return linear_scale(value, 76, 185, 101, 150)
    elif value <= 304:
        return linear_scale(value, 186, 304, 151, 200)
    elif value <= 604:
        return linear_scale(value, 305, 604, 201, 300)
    elif value <= 804:
        return linear_scale(value, 605, 804, 301, 400)
    elif value <= 1004:
        return linear_scale(value, 805, 1004, 401, 500)
    else:
        return 500  # AQI value capped at 500 for SO2

# Assuming 'clean_data1' is your DataFrame
clean_data1['AQI_O3'] = clean_data1['O3'].apply(calculate_aqi_o3)
clean_data1['AQI_PM2.5'] = clean_data1['PM2.5'].apply(calculate_aqi_pm25)
clean_data1['AQI_NO2'] = clean_data1['NO2'].apply(calculate_aqi_no2)
clean_data1['AQI_PM10'] = clean_data1['PM10'].apply(calculate_aqi_pm10)
clean_data1['AQI_CO'] = clean_data1['CO'].apply(calculate_aqi_co)
clean_data1['AQI_NO'] = clean_data1['NO'].apply(calculate_aqi_no)
clean_data1['AQI_NOx'] = clean_data1['NOx'].apply(calculate_aqi_nox)
clean_data1['AQI_SO2'] = clean_data1['SO2'].apply(calculate_aqi_so2)

# Function to determine AQI category
def classify_aqi(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

# Calculate the maximum AQI and classify
clean_data1['Max_AQI'] = clean_data1[['AQI_O3', 'AQI_PM2.5', 'AQI_NO2', 'AQI_PM10', 'AQI_CO', 'AQI_NO', 'AQI_NOx', 'AQI_SO2']].max(axis=1)
clean_data1['AQI_Category'] = clean_data1['Max_AQI'].apply(classify_aqi)

# Print the updated DataFrame
print("Updated DataFrame with AQI and Categories:")
print(clean_data1.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the 'DateLocal' column is in datetime format
clean_data1['DateLocal'] = pd.to_datetime(clean_data1['DateLocal'], errors='coerce')

# Extract month and year from the date
clean_data1['Month'] = clean_data1['DateLocal'].dt.to_period('M')
clean_data1['Year'] = clean_data1['DateLocal'].dt.year

# List of pollutants to include in AQI calculation
pollutants = ['AQI_O3', 'AQI_PM2.5', 'AQI_NO2', 'AQI_PM10']

# Create a DataFrame to store the monthly average AQI for each city
monthly_aqi = clean_data1.groupby(['location_name', 'Year', 'Month'])[pollutants].mean().reset_index()

# Number of pollutants
num_pollutants = len(pollutants)

# Create a figure with subplots for each pollutant
fig, axes = plt.subplots(num_pollutants, 1, figsize=(14, 4 * num_pollutants), sharex=True)

# Iterate over each pollutant to create a heatmap
for i, pollutant in enumerate(pollutants):
    # Pivot the DataFrame for heatmap plotting
    heatmap_data = monthly_aqi.pivot_table(index=['Year', 'Month'], columns='location_name', values=pollutant)

    # Plot heatmap for the current pollutant
    sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5, ax=axes[i])
    axes[i].set_title(f'Average Monthly {pollutant} by City')
    axes[i].set_ylabel('Year-Month')

# Set x-axis labels for the last subplot and rotate for visibility
axes[-1].set_xlabel('City')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
